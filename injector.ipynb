```python
import time

######################
INTAKE = 5           #    number before intake() starts
INJECTION_TIME = 1   #    open injector time
SPARK = 10           #    number before spark() starts 
FLYHWEEl_POINT = 15  #    flywheel mark
######################

def wheel():
  gear = []
  while True:
    a = int(input())
    gear.append(a)
    print(gear)
    if len(gear) == INTAKE:
      print("intake")
    if len(gear) == SPARK:
      print("spark")
    if len(gear) == FLYHWEEl_POINT:  
      print("start")            # flywheel mark
      break
    print("------------------")

# def intake():
#   injector = 1
#   print(injector, "fuel on")
#   time.sleep(INJECTION_TIME)
#   injector = 0
#   print(injector, "fuel off")

# def spark():
#   spark = 1
#   print(spark, "spark")


def main():
  while True:
    wheel()
    # intake()
    # spark()


if __name__ == "__main__":
  main()
```
