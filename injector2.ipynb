```python
import time


# INTAKE = number before intake() starts
# INJECTION_TIME = open injector time
# SPARK = number before spark() starts 
# FLYHWEEl_POINT = flywheel mark

rpm = 0
print(rpm)

class Wheel:

  def __init__(self, INTAKE, INJECTION_TIME, SPARK, FLYHWEEl_POINT):
    self.INTAKE = INTAKE
    self.INJECTION_TIME = INJECTION_TIME
    self.SPARK = SPARK
    self.FLYHWEEl_POINT = FLYHWEEl_POINT

  def wheel(self, INTAKE, INJECTION_TIME, SPARK, FLYHWEEl_POINT):
    gear = []
    rotation = 0
    start = time.time()
    while True:
      a = int(input())
      gear.append(a)
      print(gear)
      if len(gear) == INTAKE:
        print("intake")
      if len(gear) == SPARK:
        print("spark")
      if len(gear) == FLYHWEEl_POINT:  
        print("start")
        rotation += 1       
        break
      print("------------------")
    end = time.time()
    clock = end - start
    global rpm 
    rpm = 60 * (rotation / clock)
    print(rpm)

def change_map():
  if rpm < 5:
    a = Wheel(1, 1, 3, 5)
    print(a.wheel(1, 1, 3, 5))
  if rpm >= 5:
    b = Wheel(2, 1, 4, 6)
    print(b.wheel(2, 1, 4, 6))

def main():
    while True:
      change_map()

if __name__ == "__main__":
  main()
```
