```python
import time

# INTAKE = number before intake() starts
# INJECTION_TIME = open injector time
# SPARK = number before spark() starts 
# FLYHWEEl_POINT = flywheel mark

rpm = 0
accelerator = 0

class Wheel:

  def __init__(self, INTAKE, INJECTION_TIME, SPARK, FLYHWEEl_POINT):
    
    self.INTAKE = INTAKE
    self.INJECTION_TIME = INJECTION_TIME
    self.SPARK = SPARK
    self.FLYHWEEl_POINT = FLYHWEEl_POINT

  def wheel(self, INTAKE, INJECTION_TIME, SPARK, FLYHWEEl_POINT):

    gear = []
    rotation = 0
    start = time.time()
    while True:
      a = int(input())  # flywheel 
      gear.append(a)
      print(gear)
      if len(gear) == INTAKE:
        print("intake")
      if len(gear) == SPARK:
        print("spark")
      if len(gear) == FLYHWEEl_POINT:  
        print("start")
        rotation += 1       
        break
      print("------------------")
    end = time.time()
    clock = end - start
    global rpm 
    rpm = 60 * (rotation / clock)
    print("rpm =", rpm)
    print("acceleration =")
    global accelerator
    accelerator = int(input())  # 1-4 positions
    print("acceleration =", accelerator)

def change_map():

  if rpm < 5:
    a = Wheel(1, 1, 6, 8)
    print(a.wheel(1, 1, 6, 8))

  if accelerator == 1:
    a = Wheel(1, 2, 5, 8)
    print(a.wheel(1, 2, 5, 8))

  if accelerator == 2:
    a = Wheel(1, 3, 4, 8)
    print(a.wheel(1, 3, 4, 8))

  if accelerator == 3:
    a = Wheel(1, 4, 3, 8)
    print(a.wheel(1, 4, 3, 8))

  if accelerator == 4:
    a = Wheel(1, 5, 2, 8)
    print(a.wheel(1, 5, 2, 8))

def main():
    while True:
      change_map()

if __name__ == "__main__":
  main()
```
